# 04_aula_df_gera

```python
def gerador_reporte(c: int) -> None:
    '''
        Gera um relatório de perfilamento para o dataframe especificado
        pelo índice c na lista de dataframes.
        O relatório é salvo como um arquivo HTML.
            Parametros:
                c (int): Índice do dataframe na lista de dataframes.
            Retorna:
                None
    '''
    df = load(lista_df[c])
    display(lista_df[c])
    display(df.head())
    profile = ProfileReport(df, title="Profiling Report")
    profile.to_file(lista_df[c]+"_report.html")

    return None
```

```python
gerador_reporte(c=12)
```
