```python
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification # for example data
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2,
                            n_redundant=0, n_repeated=0, n_classes=2,
                            n_clusters_per_class=1, weights=[0.9, 0.1],
                            flip_y=0, random_state=42)
print(X.shape)
print(y.shape)
```

```python
X[:5]
```

```python
y[:5]
```

```python
unique_elements, counts = np.unique(y, return_counts=True)
print(np.asarray((unique_elements, counts)).T)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
unique_elements, counts = np.unique(y_train, return_counts=True)
print(np.asarray((unique_elements, counts)).T)
```

```python
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)

print(X_res.shape)
print(y_res.shape)
```

```python
unique_elements, counts = np.unique(y_res, return_counts=True)
print(np.asarray((unique_elements, counts)).T)
```
