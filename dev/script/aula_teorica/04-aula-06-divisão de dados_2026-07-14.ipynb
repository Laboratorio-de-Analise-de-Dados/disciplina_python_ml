```python
def train_valid_test_df_split(
        df: pd.DataFrame,
        indep: list,
        target: str = "target",
        frac_valid: float = 0.2,
        frac_test: float = 0.13) -> tuple:
    '''
    Função que divide o DataFrame em treino, teste e validação
    e separa as colunas independentes e a coluna alvo.
    Esta função divide o DataFrame em três partes: treino, teste e validação,
    e separa as colunas independentes e a coluna alvo.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame a ser dividido.
    indep : list
        Lista de colunas independentes.
    target : str, optional
        Nome da coluna alvo. Default é "target".
    frac_valid : float, optional
        Proporção do DataFrame a ser usado como validação. Default é 0.2.
    frac_test : float, optional
        Proporção do DataFrame a ser usado como teste. Default é 0.13.
    
    Returns
    -------
    tuple
        Retorna uma tupla contendo:
        - df_train: DataFrame de treino.
        - df_valid: DataFrame de validação.
        - df_test: DataFrame de teste.
        - X_train: DataFrame de treino com colunas independentes.
        - X_valid: DataFrame de validação com colunas independentes.
        - X_test: DataFrame de teste com colunas independentes.
        - y_train: Série alvo de treino.
        - y_valid: Série alvo de validação.
        - y_test: Série alvo de teste.
    '''
    index = df.sample(frac=frac_valid, random_state=42).index
    df_train = df.loc[[i for i in df.index if i not in index], :]
    df_valid = df.loc[index, :]

    index = df_train.sample(frac=frac_test, random_state=42).index
    df_train = df_10.loc[[i for i in df_train.index if i not in index], :]
    df_test = df_10.loc[index, :]

    X_train = df_train.loc[:, indep].copy()
    X_valid = df_valid.loc[:, indep].copy()
    X_test = df_test.loc[:, indep].copy()

    y_train = df_train.loc[:, target].copy()
    y_valid = df_valid.loc[:, target].copy()
    y_test = df_test.loc[:, target].copy()

    linha = "-"*10
    print(linha)
    print(f"df_train: {df_train.shape}")
    print(f"df_valid: {df_valid.shape}")
    print(f"df_test: {df_test.shape}")

    print(linha)
    print(f"X_train: {X_train.shape}")
    print(f"X_valid: {X_valid.shape}")
    print(f"X_test: {X_test.shape}")

    print(linha)
    print(f"y_train: {y_train.shape}")
    print(f"y_valid: {y_valid.shape}")
    print(f"y_test: {y_test.shape}")

    print(linha)
    print(f"Train: {df_train.shape[0]/df_10.shape[0]:.0%}")
    print(f"Valid: {df_valid.shape[0]/df_10.shape[0]:.0%}")
    print(f"Test: {df_test.shape[0]/df_10.shape[0]:.0%}")

    return (
        df_train,
        df_valid,
        df_test,
        X_train,
        X_valid,
        X_test,
        y_train,
        y_valid,
        y_test
    )
```
